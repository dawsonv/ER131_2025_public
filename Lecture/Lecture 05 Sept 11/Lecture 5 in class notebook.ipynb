```python
cds = pd.read_csv('CAISO_2017to2018_stack.csv', index_col=0)
```

```python
cds.head()
```

```python
cds_time = pd.to_datetime(cds.index)
cds_time.hour
```

```python
cds['hour']=cds_time.hour
cds.head()
```

```python
cds.pivot_table(
    values = 'MWh',
    index = 'hour',
    columns = 'Source',
    aggfunc = 'mean')
```

```python
cds['month']=cds_time.month
cds.head()
```

```python
cds.pivot_table(
    values = 'MWh',
    index = 'month',
    columns = 'Source',
    aggfunc = 'mean')
```
